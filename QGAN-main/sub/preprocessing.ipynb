```python
import pandas as pd
from IPython.display import clear_output

name = 'train_sequences.csv'
sdf = pd.read_csv(name,sep='\t')
df = pd.DataFrame()

for i in range(len(sdf)):
    r = sdf.loc[i,'sequence']
    ll = [list(r)]
    temp = pd.DataFrame(ll)
    df = df.append(temp)
    clear_output(wait=True)
```

```python
len(df)
```

```python
app = []
for i in range(len(df)):
    app = app + [0]
for j in range(512-len(df.iloc[-1])):
    df.insert(len(df.iloc[0])-1,512-j,app)
```

```python
df.fillna(0)
df
```

```python
AMINO_ACID_TO_ID = {'0': 0,
                    'A': 1,
                    'C': 2,
                    'D': 3,
                    'E': 4,
                    'F': 5,
                    'G': 6,
                    'H': 7,
                    'I': 8,
                    'K': 9,
                    'L': 10,
                    'M': 11,
                    'N': 12,
                    'P': 13,
                    'Q': 14,
                    'R': 15,
                    'S': 16,
                    'T': 17,
                    'V': 18,
                    'W': 19,
                    'Y': 20}
```

```python
for i in range(len(df)):
    temp = df.iloc[i]
    for j in range(len(temp)):
        c = temp.iloc[j]
    if c in AMINO_ACID_TO_ID:
        temp.iloc[j] = AMINO_ACID_TO_ID[c]
    elif c == float('nan'):
        temp.iloc[j] = 0
    df.iloc[i] = temp
```

```python
df.to_csv('data_numbered.csv')
```
